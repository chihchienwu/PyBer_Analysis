# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
TtlRidePerCityType = pyber_data_df.groupby(["type"]).count()["ride_id"]
TtlRidePerCityType
```

```python
# 2. Get the total drivers for each city type
TtlDriverPerCityType = city_data_df.groupby(["type"]).sum()["driver_count"]
TtlDriverPerCityType
```

```python
#  3. Get the total amount of fares for each city type
TtlFarePerCityType = pyber_data_df.groupby(["type"]).sum()["fare"]
TtlFarePerCityType
```

```python
#  4. Get the average fare per ride for each city type. 
AvgFarePerCityType = pyber_data_df.groupby(["type"]).mean()["fare"]
AvgFarePerCityType
```

```python
# 5. Get the average fare per driver for each city type. 
NewTable = pd.merge(TtlDriverPerCityType, TtlFarePerCityType, how="left", on=["type", "type"])
NewTable["AvgFare"] = NewTable["fare"]/NewTable["driver_count"]
AvgFarePerDriver = NewTable.groupby(["type"]).mean()["AvgFare"]
AvgFarePerDriver 
```

```python
#  6. Create a PyBer summary DataFrame. 
Delv1_summary_df = pd.DataFrame({
          "Total Ride" : TtlRidePerCityType,
          "Total Driver": TtlDriverPerCityType,
          "Total Fares": TtlFarePerCityType,
          "Average Fare per Ride": AvgFarePerCityType,
          "Average Fare per Driver": AvgFarePerDriver})
Delv1_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
Delv1_summary_df.index.name = None
```

```python
#  8. Format the columns.
Delv1_summary_df["	Total Ride"] = Delv1_summary_df["Total Ride"].map("{:,.0f}".format)
Delv1_summary_df["Total Driver"] = Delv1_summary_df["Total Driver"].map("{:,.0f}".format)
Delv1_summary_df["Total Fares"] = Delv1_summary_df["Total Fares"].astype(float)
Delv1_summary_df["Total Fares"] = Delv1_summary_df["Total Fares"].map("${:,.2f}".format)
Delv1_summary_df["Average Fare per Ride"] = Delv1_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
Delv1_summary_df["Average Fare per Driver"] = Delv1_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
Delv1_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

NewFrame_Df = pyber_data_df.groupby(["date","type"]).sum()["fare"]
NewFrame_Df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()

NewFrame_Df = NewFrame_Df.reset_index()
NewFrame_Df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

PivotFrame_df = NewFrame_Df.pivot(index='date', columns='type', values='fare')
PivotFrame_df 
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

LocFrame_df = PivotFrame_df.loc['2019-01-01' : '2019-04-29']
LocFrame_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

LocFrame_df.index = pd.to_datetime(LocFrame_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()

LocFrame_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

FinalFrame_df = LocFrame_df.resample('W', convention='end').sum()
FinalFrame_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
# Use the graph style fivethirtyeight.
import matplotlib.dates as mdates
from matplotlib import style
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12.5, 5))
ax.plot(FinalFrame_df.index, FinalFrame_df["Rural"],label='Rural',linewidth=2)
ax.plot(FinalFrame_df.index, FinalFrame_df["Suburban"],label='Suburban',linewidth=2)
ax.plot(FinalFrame_df.index, FinalFrame_df["Urban"],label='Urban',linewidth=2)

ax.set_title('Total Fare by City Type',fontsize = 15,fontweight=100)
ax.set_ylabel('Fare ($USD)',fontsize = 15)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b')) 
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.legend(title='Type',loc='center')

plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()

```
